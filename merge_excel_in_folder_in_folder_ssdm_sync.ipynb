```python
import os
import glob2
import pandas as pd

folder_path = r"C:\Users\Aiman\Downloads\ssdm data ppd jb (mac-jun 2023)\raw"
output_folder = r"C:\Users\Aiman\Downloads\ssdm data ppd jb (mac-jun 2023)\appended ssdm"

subfolders = glob2.glob(os.path.join(folder_path, "*"))

for subfolder in subfolders:
    subfolder_name = os.path.basename(subfolder)
    output_file = os.path.join(output_folder, subfolder_name + ".xlsx")

    filenames = glob2.glob(os.path.join(subfolder, "*.xlsx"))

    outputcsv = pd.DataFrame()

    for file in filenames:
        df = pd.read_excel(file)
        Synced = 0
        df['Synced'] = Synced
        outputcsv = outputcsv.append(df, ignore_index=True)
    
    outputcsv.drop(outputcsv.columns[[1, 4, 5, 6]], axis=1, inplace=True)

    outputcsv.to_csv(output_file, index=False)

    print("Final Excel sheet generated and saved to:", subfolder_name)
```
