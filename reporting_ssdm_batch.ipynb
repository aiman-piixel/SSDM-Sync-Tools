```python
import os
import glob2
import pandas as pd
import numpy as np
import xlsxwriter
import shutil

ssdm1 = [#SUKARELA MEMBANTU GURU
        'MEMBANTU GURU JIKA ADA MAJLIS',
        'MENOLONG DAN MENGAWAL KELAS SEMASA GURU TIADA',
        'MEMBANTU GURU MENGAMBIL KEHADIRAN PELAJAR',
        'MEMBANTU GURU JIKA MEMERLUKAN BANTUAN',
        'MEMBANTU MENGANGKAT BUKU/BARANG GURU']
ssdm2 = [#MENZAHIRKAN MANA-MANA AMALAN BAIK
        'DATANG AWAL KE SEKOLAH',
        'BERDISIPLIN',
        'BERCAKAP DENGAN SOPAN',
        'BERPAKAIAN KEMAS',
        'MENYAPA/MEMBERI SALAM KEPADA GURU']
ssdm3 = [#MEMBANTU MENGEMAS MANA-MANA AMALAN BAIK
        'MENYUSUN KEMAS MEJA',
        'MEMBUANG DEBU/HABUK DI CERMIN/MEJA/DINDING',
        'MEMBERSIHKAN PAPAN HITAM',
        'MENGEMOP BILIK',
        'MENYAPU BILIK-BILIK KHAS']
ssdm4 = [#MEMBERSIHKAN KAWASAN/BILIK KHAS
        'MENYUSUN MEJA/PERABOT',
        'MENGHIAS BILIK',
        'MEMBERSIHKAN MEJA/PERABOT YANG BERDEBU',
        'MOP LANTAI',
        'MENYAPU SAMPAH-SARAP DI SEKELILING KELAS']
ssdm5 = [#MENYERTAI AKTIVITI DIBAWAH BIMBINGAN DAN KAUNSELING
        'MUDAH BERINTERAKSI ATAU BERGAUL',
        'MENGHORMATI PENDAPAT ORANG LAIN',
        'BIJAK BERKOMUNIKASI UNTUK MENDAPATKAN MAKLUMAT',
        'MENJADI FASILATATOR']
ssdm6 = [#SUKARELA MELAPOR KES SALAH LAKU
        'MELAPOR KES MASALAH SOSIAL YANG SERIUS',
        'MELAPOR KES MEMBAWA/MENGGUNAKAN BARANG TERLARANG',
        'MELAPOR KES TIDAK BERDISIPLIN',
        'MELAPOR KES LEWAT KE SEKOLAH',
        'MELAPOR KES PONTENG SEKOLAH/KELAS']
ssdm7 = [#MEMBANTU MELAKSANA KERJA-KERJA LANDSKAP/TAMAN
        'MENYUSUN DAN MENCANTIK KAN KAWASAN TAMAN/LANDSKAP',
        'MEMOTONG RUMPUT YANG PANJANG',
        'MEMBANTU MENYIRAM POKOK BUNGA',
        'MENGUTIP SAMPAH-SARAP YANG ADA',
        'MEMBANTU MENYAPU DAUN-DAUN KERING']
ssdm8 = [#KERJA-KERJA MENGECAT
        'MELAKUKAN DENGAN PENUH SEMANGAT',
        'SUKA MENOLONG ORANG LAIN',
        'MEMBANTU MENGEMAS SEMULA BARANG YANG DIGUNAKAN',
        'BEKERJASAMA MENGECAT DINDING',
        'MENUNJUKKAN KREATIVITI']
all_ssdm = ssdm1+ssdm2+ssdm3+ssdm4+ssdm5+ssdm6+ssdm7+ssdm8

#rename accordingly to OS
folder_path = "/Users/piixel/Desktop/ssdm terengganu (1jun- 31aug)"
export_path = "/Users/piixel/Desktop/ssdm terengganu (1jun- 31aug) - report"

if os.path.exists(export_path):
    shutil.rmtree(export_path)  # Remove the existing directory and its contents
os.makedirs(export_path)

filenames = glob2.glob(folder_path + "/*.xlsx")
filenamesArr = np.array(filenames)

for i in filenamesArr:

    df = pd.read_excel(i)

    # seperating dataframe based on months and category - adjust this part based on your need
    jun_df = df[df.apply(lambda row: 'Jun' in str(row), axis=1)]
    jul_df = df[df.apply(lambda row: 'Jul' in str(row), axis=1)]
    aug_df = df[df.apply(lambda row: 'Aug' in str(row), axis=1)] 

    jun_df.reset_index(drop=True, inplace=True)
    jun1_df = jun_df[jun_df.apply(lambda row: any(string in str(row) for string in ssdm1), axis=1)]
    jun2_df = jun_df[jun_df.apply(lambda row: any(string in str(row) for string in ssdm2), axis=1)]
    jun3_df = jun_df[jun_df.apply(lambda row: any(string in str(row) for string in ssdm3), axis=1)]
    jun4_df = jun_df[jun_df.apply(lambda row: any(string in str(row) for string in ssdm4), axis=1)]
    jun5_df = jun_df[jun_df.apply(lambda row: any(string in str(row) for string in ssdm5), axis=1)]
    jun6_df = jun_df[jun_df.apply(lambda row: any(string in str(row) for string in ssdm6), axis=1)]
    jun7_df = jun_df[jun_df.apply(lambda row: any(string in str(row) for string in ssdm7), axis=1)]
    jun8_df = jun_df[jun_df.apply(lambda row: any(string in str(row) for string in ssdm8), axis=1)]
    junAll_df = jun_df[jun_df.apply(lambda row: any(string in str(row) for string in all_ssdm), axis=1)]

    jul_df.reset_index(drop=True, inplace=True)
    jul1_df = jul_df[jul_df.apply(lambda row: any(string in str(row) for string in ssdm1), axis=1)]
    jul2_df = jul_df[jul_df.apply(lambda row: any(string in str(row) for string in ssdm2), axis=1)]
    jul3_df = jul_df[jul_df.apply(lambda row: any(string in str(row) for string in ssdm3), axis=1)]
    jul4_df = jul_df[jul_df.apply(lambda row: any(string in str(row) for string in ssdm4), axis=1)]
    jul5_df = jul_df[jul_df.apply(lambda row: any(string in str(row) for string in ssdm5), axis=1)]
    jul6_df = jul_df[jul_df.apply(lambda row: any(string in str(row) for string in ssdm6), axis=1)]
    jul7_df = jul_df[jul_df.apply(lambda row: any(string in str(row) for string in ssdm7), axis=1)]
    jul8_df = jul_df[jul_df.apply(lambda row: any(string in str(row) for string in ssdm8), axis=1)]
    julAll_df = jul_df[jul_df.apply(lambda row: any(string in str(row) for string in all_ssdm), axis=1)]

    aug_df.reset_index(drop=True, inplace=True)
    aug1_df = aug_df[aug_df.apply(lambda row: any(string in str(row) for string in ssdm1), axis=1)]
    aug2_df = aug_df[aug_df.apply(lambda row: any(string in str(row) for string in ssdm2), axis=1)]
    aug3_df = aug_df[aug_df.apply(lambda row: any(string in str(row) for string in ssdm3), axis=1)]
    aug4_df = aug_df[aug_df.apply(lambda row: any(string in str(row) for string in ssdm4), axis=1)]
    aug5_df = aug_df[aug_df.apply(lambda row: any(string in str(row) for string in ssdm5), axis=1)]
    aug6_df = aug_df[aug_df.apply(lambda row: any(string in str(row) for string in ssdm6), axis=1)]
    aug7_df = aug_df[aug_df.apply(lambda row: any(string in str(row) for string in ssdm7), axis=1)]
    aug8_df = aug_df[aug_df.apply(lambda row: any(string in str(row) for string in ssdm8), axis=1)]
    augAll_df = aug_df[aug_df.apply(lambda row: any(string in str(row) for string in all_ssdm), axis=1)]


    junData = [['SUKARELA MEMBANTU GURU', jun1_df.shape[0]], 
               ['MENZAHIRKAN MANA-MANA AMALAN BAIK', jun2_df.shape[0]],
               ['MEMBANTU MENGEMAS MANA-MANA AMALAN BAIK', jun3_df.shape[0]],
               ['MEMBERSIHKAN KAWASAN/BILIK KHAS', jun4_df.shape[0]],
               ['MENYERTAI AKTIVITI DIBAWAH BIMBINGAN DAN KAUNSELING', jun5_df.shape[0]],
               ['SUKARELA MELAPOR KES SALAH LAKU', jun6_df.shape[0]],
               ['MEMBANTU MELAKSANA KERJA-KERJA LANDSKAP/TAMAN', jun7_df.shape[0]],
               ['KERJA-KERJA MENGECAT', jun8_df.shape[0]],
               ['JUMLAH AMALAN BAIK', junAll_df.shape[0]]
              ]
    julData = [['SUKARELA MEMBANTU GURU', jul1_df.shape[0]], 
               ['MENZAHIRKAN MANA-MANA AMALAN BAIK', jul2_df.shape[0]],
               ['MEMBANTU MENGEMAS MANA-MANA AMALAN BAIK', jul3_df.shape[0]],
               ['MEMBERSIHKAN KAWASAN/BILIK KHAS', jul4_df.shape[0]],
               ['MENYERTAI AKTIVITI DIBAWAH BIMBINGAN DAN KAUNSELING', jul5_df.shape[0]],
               ['SUKARELA MELAPOR KES SALAH LAKU', jul6_df.shape[0]],
               ['MEMBANTU MELAKSANA KERJA-KERJA LANDSKAP/TAMAN', jul7_df.shape[0]],
               ['KERJA-KERJA MENGECAT', jul8_df.shape[0]],
               ['JUMLAH AMALAN BAIK', julAll_df.shape[0]]
              ]
    augData = [['SUKARELA MEMBANTU GURU', aug1_df.shape[0]], 
               ['MENZAHIRKAN MANA-MANA AMALAN BAIK', aug2_df.shape[0]],
               ['MEMBANTU MENGEMAS MANA-MANA AMALAN BAIK', aug3_df.shape[0]],
               ['MEMBERSIHKAN KAWASAN/BILIK KHAS', aug4_df.shape[0]],
               ['MENYERTAI AKTIVITI DIBAWAH BIMBINGAN DAN KAUNSELING', aug5_df.shape[0]],
               ['SUKARELA MELAPOR KES SALAH LAKU', aug6_df.shape[0]],
               ['MEMBANTU MELAKSANA KERJA-KERJA LANDSKAP/TAMAN', aug7_df.shape[0]],
               ['KERJA-KERJA MENGECAT', aug8_df.shape[0]],
               ['JUMLAH AMALAN BAIK', augAll_df.shape[0]]
              ]

    activity_counts = {}
    for data in [junData, julData, augData]:
        for activity, count in data:
            if activity in activity_counts:
                activity_counts[activity] += count
            else:
                activity_counts[activity] = count

    totalData = [[activity, count] for activity, count in activity_counts.items()]


    allRecord = pd.DataFrame(totalData, columns=['KATEGORI AMALAN BAIK', 'JUMLAH SEMUA'])
    junRecord = pd.DataFrame(junData, columns=['KATEGORI AMALAN BAIK', 'JUMLAH JUN'])
    julRecord = pd.DataFrame(julData, columns=['KATEGORI AMALAN BAIK', 'JUMLAH JULAI'])
    augRecord = pd.DataFrame(augData, columns=['KATEGORI AMALAN BAIK', 'JUMLAH OGOS'])

    fileName = i
    exportName = fileName.replace("/Users/piixel/Desktop/ssdm terengganu (1jun- 31aug)","")
    writer = pd.ExcelWriter(export_path + exportName, engine='xlsxwriter')

    dataframes = {
        'Jumlah' : allRecord,
        'Jun' : junRecord,
        'Jul' : julRecord,
        'Aug' : augRecord
    }

    for sheet_name, export in dataframes.items():
        export.to_excel(writer, sheet_name=sheet_name, index=False)
        worksheet = writer.sheets[sheet_name]
        worksheet.set_column('A:A', 70)
        worksheet.set_column('B:B', 15)

    print('Report has been generated for ' +i)
    writer.save()
```
