```python
import os
import glob2
import pandas as pd

# Input and output folder paths
folder_path = r"C:\Users\Aiman\Desktop\ssdm terengganu 1jun-31aug"
output_folder = r"C:\Users\Aiman\Desktop\ssdm terengganu 1jun-31aug\appended"

# Get a list of subfolders in the input folder
subfolders = glob2.glob(os.path.join(folder_path, "*"))

# Iterate through each subfolder
for subfolder in subfolders:
    subfolder_name = os.path.basename(subfolder)
    output_file = os.path.join(output_folder, subfolder_name + ".xlsx")  # Change extension to CSV

    # Get a list of Excel files in the current subfolder
    filenames = glob2.glob(os.path.join(subfolder, "*.xlsx"))
    
    if len(filenames) == 0:
        print("Skipping empty folder:", subfolder_name)
        continue

    output_excel = pd.DataFrame()

    # Iterate through each Excel file
    for file in filenames:
        df = pd.read_excel(file)
        df['Synced'] = 0  # Add a 'Synced' column filled with zeros
        output_excel = pd.concat([output_excel, df], ignore_index=True)  # Use pd.concat() for concatenation

    # Drop specific columns from the DataFrame
    columns_to_drop = [1, 4, 5, 6]
    output_excel.drop(output_excel.columns[columns_to_drop], axis=1, inplace=True)

    # Save the combined DataFrame to a CSV file
    output_excel.to_excel(output_file, index=False)

    print("Final Excel file generated and saved to:", subfolder_name)
```
